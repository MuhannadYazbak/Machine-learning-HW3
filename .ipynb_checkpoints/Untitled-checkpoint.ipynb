# CountyData Exploration

```python
land = pd.read_csv("data/CountyData/5296US_landarea.csv")
display(land.head())
```

```python
land.info()
```

```python
water = pd.read_csv("data/CountyData/5296US_waterarea.csv")
display(water.head())
```

```python
water.info()
```

```python
pop = pd.read_csv("data/CountyData/5296US_pop.csv")
display(pop.head())
```

```python
pop.info()
```

```python
counties = pd.merge(land, water)
counties = pd.merge(counties, pop)
display(counties.head())
```

```python
for col in counties.columns:
    if counties[col].dtype == 'object':
        counties[col+"Code"] = pd.factorize(counties[col])[0]
        
display(counties.head())
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(counties.corr(), annot=True)
plt.title("Counties features correlation matrix", fontsize=16, color="cornflowerblue")
plt.show()
```

```python
plt.figure(figsize=(20,20))
sns.pairplot(data=counties, corner=True)
plt.show()
```

```python
plt.figure(figsize=(15,15))
sns.countplot(x='State', data=counties)
plt.ylabel("Number of Counties")
plt.xlabel("State")
plt.title("Number of counties in each State", fontsize=16, color='cornflowerblue')
plt.xticks(rotation = 90)
plt.show()
```

```python
counties.isna().sum()
```

```python
display(counties['Water Area'].unique())
```

```python
counties.replace(np.nan, 0, inplace= True)
display(counties.isna().sum())
```

# Life Expectancy Data Exploration

```python
life = pd.read_csv("data/LifeExpectancyData/IHME_USA_LIFE_EXPECTANCY_1987_2007_Y2011M06D16.csv")
display(life.head())
```

```python
life.info()
```

```python
plt.figure(figsize=(20,20))
sns.set_theme(style="whitegrid")

# Draw a nested barplot by species and sex
g = sns.catplot(
    data=life, kind="bar",
    x="State", y="Rank (male)",
    ci="sd", palette="dark", alpha=.6, height=6
)
g.despine(left=True)
plt.xticks(rotation = 90)
g.set_axis_labels("", "Rank (male)")
plt.show()
```

```python
states = life[life['State'].isin(['CALIFORNIA', 'FLORIDA', 'SOUTH DAKOTA','WYOMING'])]
display(states.head())
```

```python
states.info()
```

```python
plt.figure(figsize=(20,20))
sns.set_theme(style="whitegrid")

# Draw a nested barplot by species and sex
g = sns.catplot(
    data=states, kind="bar",
    x="State", y="Rank (male)", hue='Year',
    ci="sd", palette="dark", alpha=.6, height=6
)
g.despine(left=True)
plt.xticks(rotation = 90)
g.set_axis_labels("", "Rank (male)")
plt.show()
```

```python
sigma = states.cov()
sigma
```

```python
eigen = np.linalg.eig(sigma)
eigen
```

```python
eigenvalues = pd.DataFrame(eigen[0].reshape(1,16), columns=sigma.columns, index=['Value'])
eigenvalues
```

```python
eigenvectors = pd.DataFrame(eigen[1], columns=sigma.columns, index=sigma.columns)
eigenvectors
```

```python
PCs = states @ eigenvectors # @ - matrix multliplication
```

```python
from sklearn.decomposition import PCA
PC = PCA(n_components=2).fit(sigma)
```

```python
PC.explained_variance_ # eigenvalues
```

```python
PC.components_ # eigenvectors
```
