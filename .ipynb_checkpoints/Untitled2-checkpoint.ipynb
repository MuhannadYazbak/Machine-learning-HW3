```python
water = pd.read_csv("data/CountyData/5296US_waterarea.csv")
display(water.info())
display(water.head())
```

```python
print(water.isna().sum())
```

Water area has 353 missing or null values, which in order to impute we will use KNNImputer with 10 nearest neighbors to keep the data as real as it can be, but in order to be able to use KNN we must factorize the object typed features into numeric features, so we will create a copy of the dataframe and then replace the original water area column with the imputed one.

```python
water['Water Area'].unique()
```

The only missing value is nan 

```python
cwater = water.copy()
cwater
```

```python
for col in cwater.columns:
    if cwater[col].dtype == 'object':
        cwater[col] = pd.factorize(cwater[col])[0]
cwater
```

```python
knn = KNNImputer(n_neighbors= 10).fit(cwater)
cwater = pd.DataFrame(knn.transform(cwater), columns=water.columns)
```

```python
cwater.isna().sum()
```

```python
water['Water Area'] = cwater['Water Area']
display(water.info())
```

```python
water.isna().sum()
```

```python
# lower case headers
countyData.columns = countyData.columns.str.lower()

# lower case all string values
countyData = countyData.apply(lambda s: s.lower() if type(s) == str else s)

display(countyData.head(10))
```

```python
dow = pd.read_csv("data/EconomicData/Dow Jones Industrial Average Historical Data 2.csv")
display(dow.info())
display(dow)
```

```python
dow.isna().sum()
```

```python
dow['Date'] = pd.to_datetime(dow['Date'])
dow['day'] = dow['Date'].dt.strftime('%d').astype(int)
dow['month'] = dow['Date'].dt.strftime('%m').astype(int)
dow['year'] = dow['Date'].dt.strftime('%Y').astype(int)
dow = dow[['day','month','year', 'Price','Open', 'High', 'Low', 'Vol.', 'Change %']]
dow
```

```python
dowObject = dow.select_dtypes(exclude='int')
dowObject
```

```python
dowObject.replace('-',0,inplace=True)
for col in dowObject.columns:
    if col == 'Vol.':
        dowObject[col] = dowObject[col].replace({'M':''}, regex=True).apply(pd.to_numeric,1)
    elif col == 'Change %':
        dowObject[col] = dowObject[col].replace({'%':''}, regex=True).apply(pd.to_numeric,1)
    else:
        dowObject[col] = dowObject[col].replace({',':''},regex=True).apply(pd.to_numeric,1)
```

```python
dowObject
```

```python
dowObject.info()
```

```python
dow = pd.concat([dow[['day','month','year']], dowObject], axis=1)
dow
```

```python
plt.figure(figsize=(15,15))
sns.scatterplot(data=dow, x='year', y='Price', hue='month')
plt.title("Dow Jones: year Vs. Price", fontsize=15, color='red')
plt.xticks(rotation = 90)
plt.show()
```
