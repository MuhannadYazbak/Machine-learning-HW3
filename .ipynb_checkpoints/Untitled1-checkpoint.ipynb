# Building dataframes, Exploration and Visualization

## CountyData

```python
# readme to know what we are expecting to have
cdreadme = pd.read_table("data/CountyData/5296readme.txt")
print(cdreadme)
```

We expect 3 dataframes (landarea, waterarea and population)

```python
# land area
land = pd.read_csv("data/CountyData/5296US_landarea.csv")
display(land.info())

# water area
water = pd.read_csv("data/CountyData/5296US_waterarea.csv")
display(water.info())

# population
pop = pd.read_csv("data/CountyData/5296US_pop.csv")
display(pop.info())
```

We can see that the 3 dataframes have the same number of entries (rows) and have features in common so to make it easier we will merge the 3 into one dataframe named countyData, and explore further.

```python
countyData = pd.merge(land,pd.merge(water, pop))
display(countyData.head())
```

```python
# check for null or missing values to impute
countyData.isna().sum()
```

only water area have 183 missing values among the 3111 entries, so first of all we should explore and tell which value represents these missing values and accordingly to decide how to impute it.

```python
countyData['Water Area'].unique()
```

the missing values have np.nan, and since no further instruction is given we will impute all missing value to be 0

```python
countyData.replace(np.nan, 0, inplace= True)
display(countyData.isna().sum())
```

in order to make it easier to make further merges and since python is case sensitive we will :

1- lower case all header names and all String feature values.

2- add additional feature of the string feature with Code suffix and factorize into numeric values (will help visualizing correlation matrix)

```python
# lower case headers
countyData.columns= countyData.columns.str.lower()
countyData.columns
```

```python
# lower case String columns
countyData = countyData.applymap(lambda s:s.lower() if type(s) == str else s)
display(countyData.head(10))
```

```python
countyData.info()
```

```python
# add numeric features
for col in countyData.columns:
    if countyData[col].dtype == 'object':
        countyData[col+'Code'] = pd.factorize(countyData[col])[0]
        
        
display(countyData.head(10))
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(countyData.corr(), annot=True, cmap="YlGnBu")
plt.title("CountyData features correlation matrix", fontsize=16, color='cornflowerblue')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.countplot(x='state', data=countyData)
plt.title("Counties number in each state", fontsize=14, color='cornflowerblue')
plt.xticks(rotation= 90)
plt.show()
```

```python
# dataframe showing sum of landarea,waterarea and population in each state
states = countyData.groupby('state').sum()
states = countyData[['land area', 'water area', 'population']]
display(states)
```

We read the 3 dataframes in CoutyData, merged into one dataframe, imputed missing values with 0, added numeric features representing the String features and visualized correlation matrix but none interesting relations to further explore.

We visualized the number of counties in each state.

We grouped by state eliminated none related sums of fips and codes then showed a table with total landarea,waterarea and population in each state.

Now gdp is fixed, but also couty1 dataframe looks like it need some cleaning as well (hence the unnamed headers)

```python
display(c1.head(15))

display(c1.tail(15))
```

```python
c1 = pd.read_csv("data/EconomicData/county1.csv", skipfooter=13,skiprows=5)
c1.head(10)
```

```python
c1.tail(15)
```

```python
c1.drop(columns=['Unnamed: 5'],inplace=True)
c1.head()
```

```python
c1.dropna(thresh=3, inplace=True)
c1.reset_index(drop=True, inplace=True)
c1.info()
```

```python
c1.columns = ['state', 'MHS1999', 'MHS1989', 'MHS1979', 'MHS1969']
c1.info()
```

```python
for col in c1.columns:
    if col == 'state':
        continue
    c1[col].replace(',','', regex=True, inplace=True)
    c1[col].replace(' ',0, regex=True,inplace=True)
    c1[col].replace('NA', 0, regex=True, inplace=True)
    c1[col].replace('(a)', 0, regex=True, inplace=True)
    c1[col] = pd.to_numeric(c1[col])
c1.info()
```

```python
c1.fillna(0, inplace=True)
c1.info()
```

```python
display(c1.head(15))
display(c1.tail(15))
```
